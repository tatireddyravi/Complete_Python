- The statement **import matplotlib.pyplot as plt** is a common way to import the pyplot module from the Matplotlib library. This module provides a collection of functions that allow you to create a variety of plots and visualizations with a simple interface similar to MATLAB.

```python
import matplotlib.pyplot as plt
```

```python
# creating some sample data
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
```

```python
# creating a simple lineplot
plt.plot(x,y1)
# plt.plot will show line plot by default

'''if we press (shift + tab) key inside the plot function,
then we can see the documentation of the plt.plot()'''
```

```python
# creating a simple lineplot
plt.plot(x,y1)
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Simple Line Plot')
# Display the plot
plt.show()
```

- In above example, **plt.plot()** is used to create a line plot, **plt.xlabel()** and **plt.ylabel()** add labels to the axes, and **plt.title()** sets the title of the plot. Finally, **plt.show()** is called to display the plot on the screen.

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
# Drawing two plots in a single diagram
y2=[8,5,9,12,8]
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
# Drawing two plots in a single diagram
y2=[8,5,9,12,8]
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
# add lagends
plt.legend('y1','y2')
plt.show()
```

#### Examples od format string 

**'b'** - blue markers with default shape

**'or'** - red circles'-g' # green solid line

**'^k:'** - black triangle_up markers connected by a dotted line

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
plt.plot(x,y1,'--k',label='y1_data')
plt.plot(x,y2,'b-*',label='y2_data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend() 
plt.show()
```

#### pass color and linestyle as an argument

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
plt.plot(x,y1,color='b',linestyle=':',label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',label='y2_data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend() 
plt.show()
```

#### adding marker

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
plt.plot(x,y1,color='b',linestyle=':',marker='3',label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',marker='*',label='y2_data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend() 
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
y3=[10,9,15,16,17]
plt.plot(x,y1,color='c',linestyle=':',marker='3',label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',marker='*',label='y2_data')
plt.plot(x,y3,marker='H',label='y3_data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend() 
plt.show()
```

#### adjust thickness of line using linewidth argument

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
y3=[10,9,15,16,17]
plt.plot(x,y1,color='c',linestyle=':',marker='3',linewidth=2,label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',marker='*',linewidth=3,label='y2_data')
plt.plot(x,y3,marker='H',label='y3_data',linewidth=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend() 
plt.show()
```

- The **plt.tight_layout()** function in Matplotlib is used to automatically adjust the spacing between subplots or plot elements in a figure. It helps ensure that all plot components are properly visible and prevent overlap, making your plots more visually appealing, especially when you have multiple subplots.

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
y3=[10,9,15,16,17]
plt.plot(x,y1,color='c',linestyle=':',marker='3',linewidth=2,label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',marker='*',linewidth=3,label='y2_data')
plt.plot(x,y3,marker='H',label='y3_data',linewidth=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Create two subplots
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [4, 5, 6])
plt.title('Plot 1')

plt.subplot(1, 2, 2)
plt.plot([3, 2, 1], [6, 5, 4])
plt.title('Plot 2')

plt.show()
```

- The **plt.grid()** function in Matplotlib is used to add grid lines to a plot. Grid lines can be helpful in providing a visual reference for the data points and their positions, making it easier to interpret the plot.

```python
x=[1,2,3,4,5]
y1=[5,3,7,9,6]
y2=[8,5,9,12,8]
y3=[10,9,15,16,17]
plt.plot(x,y1,color='c',linestyle=':',marker='3',linewidth=2,label='y1_data')
plt.plot(x,y2,color='g',linestyle='-.',marker='*',linewidth=3,label='y2_data')
plt.plot(x,y3,marker='H',label='y3_data',linewidth=4)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

- **plt.style** is a Matplotlib feature that allows you to change the visual appearance of your plots by applying different styles. Styles are sets of predefined parameters that control various aspects of your plots, such as colors, line styles, grid styles, font sizes, and more. Matplotlib comes with a variety of built-in styles that you can easily apply to your plots to achieve a consistent and appealing look.

```python
import matplotlib.pyplot as plt

# Available styles
print(plt.style.available)

```

```python
# Choose a style
plt.style.use('seaborn-darkgrid')

# Create sample data
x = [1, 2, 3, 4, 5]
y = [5, 3, 8, 2, 6]

# Create a simple line plot
plt.plot(x, y)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Styled Line Plot')

# Display the plot
plt.show()
```

```python
plt.style.use('classic')
x = [1, 2, 3, 4, 5]
y = [5, 3, 8, 2, 6]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Styled Line Plot')
plt.show()
# By changing the plot style we can change the visual appearence of plot.
```

### Pie chart

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
plt.pie(sizes,labels)
```

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
plt.pie(sizes,labels=labels)
```

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
colors = ['red', 'yellow', 'orange', 'purple']
plt.pie(sizes,labels=labels,colors=colors)
```

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
colors = ['red', 'yellow', 'orange', 'purple']
explode = (0.1, 0, 0, 0)  # Explode the first slice (Apples)
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
         shadow=True, startangle=70) 
```

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [30, 25, 20, 25]
colors = ['red', 'yellow', 'orange', 'purple']
explode = (0.1, 0, 0, 0)  # Explode the first slice (Apples)
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=70) 
#here '%1.1f%%' is format string and .1 is for number of decimal points
```

- we use the **plt.pie()** function to create a pie chart. 

- The **labels** parameter specifies the labels for each slice, the sizes parameter specifies the sizes of each slice, and the **colors** parameter sets the colors for the slices. The **explode** parameter determines how much each slice should be exploded from the center. The **autopct** parameter adds percentage labels to each slice. The **shadow** parameter adds a shadow effect, and startangle sets the starting angle for the first slice.
